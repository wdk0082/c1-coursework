```python
# Create synthetic data

# 5 dim random x, polynomial f(x), and noise
import numpy as np
import pandas as pd


def create_synthetic_data(n_samples=1000, noise_std=0.1, random_seed=42):
    np.random.seed(random_seed)
    poly_coeffs = np.random.randn(5)
    X = np.random.randn(n_samples, 5)
    y = (
        poly_coeffs[0]
        + poly_coeffs[1] * X[:, 0]
        + poly_coeffs[2] * X[:, 1] ** 2
        + poly_coeffs[3] * X[:, 2] ** 3
        + poly_coeffs[4] * X[:, 3] ** 4
        + noise_std * np.random.randn(n_samples)
    )
    print("Polynomial coefficients:", poly_coeffs)
    print("X shape:", X.shape)
    print("y shape:", y.shape)
    # save as .npz
    np.savez(
        "backend_data_test.npz",
        X=X,
        y=y,
    )

create_synthetic_data(
    n_samples=1000,
    noise_std=0.1,
    random_seed=42,
)
```
